```python
# -*- coding: cp1252 -*-
from textblob import TextBlob
import csv
import re
import pandas as pd

base_dir = '../../../../data/Tweets/all_tweets_with_extractions_from_spade/'
file_input_name = 'merged_relations_-1_selected_columns.csv'
input_name_prefix = 'twitter_mobile_payments'
output_file = 'merged_relations_-1_selected_columns_with_sentiments.csv'
DATA_SET = 'twitter'

def read_df_rel(based_dir, file_input_name):
    file_input = based_dir + file_input_name    
    ff = open(file_input)
    delim=","
    df = pd.read_csv(file_input,delimiter=delim,header=0)        
    return df

df = read_df_rel(based_dir, file_input_name)
```
