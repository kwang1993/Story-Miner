```python
#step 1
%matplotlib inline 

import pandas as pd
from collections import defaultdict
import re
import numpy as np
import sys
import seaborn as sns
import matplotlib.pyplot as plt

pal = sns.color_palette()


sys.path.insert(0, '../../base_codes')
sys.path.insert(0, '../../data_specific_codes')
sys.path.insert(0, '../../utility_codes')


from RE_init import *
from main_functions import *
from utility_functions import *

def read_df_rel():
    #based_dir = '../../../../data/GoodReads/'
    based_dir = '../../../../data/Tweets/'
    file_input_name = 'merged.csv'
    file_input = based_dir + file_input_name    
    ff = open(file_input)
    delim=","
    df = pd.read_csv(file_input,delimiter=delim,header=0)        
    return df

df_rels = read_df_rel()
#rename a column
df_rels.rename(columns = {'Replaced Version of Main Tweet':'text'}, inplace = True)
#take unique tweets
df_unique = df_rels.drop_duplicates('text', inplace = False)
df_unique.reset_index(inplace=True)
#make the text to be lower case
df_unique = df_unique.apply(lambda x: x.astype(str).str.lower())
```

```python
df_sample = df_unique.sample(100)
```

```python
#import banks
df_banks = pd.read_csv('../../../../data/Tweets/Evaluation/banks_freq.csv',delimiter=",",header=0)     
#import merchants 
df_merchants = pd.read_csv('../../../../data/Tweets/Evaluation/merchants_freq.csv', delimiter=",", header=0)
```

```python
df_mobile_payments = pd.DataFrame(["applepay","googlewallet","samsungpay"],columns=["mp"])
```

```python
dict_3categ_tweet_ind = defaultdict(list)
```

```python
import time
start_time = time.time()
for ind, row in df_unique.iterrows():#df_sample.iterrows():#
    text = row["text"]
    has_bank, has_merch, has_mp = False,False,False
    for tmp,b in df_banks.iterrows():
        if b["Bank"] in text:
            has_bank = True
            break
    for tmp, merch in df_merchants.iterrows():
        if merch["Merchant"] in text:
            has_merch = True
            break
    for tmp, mp in df_mobile_payments.iterrows():
        if mp["mp"] in text:
            has_mp = True
            break
    if has_mp and has_merch:
        dict_3categ_tweet_ind["mp_merch"].append(ind)
    if has_merch and has_bank:
        dict_3categ_tweet_ind["bank_merch"].append(ind)
    if has_bank and has_mp:
        dict_3categ_tweet_ind["bank_mp"].append(ind)
        
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
def is_entity_present(sent, entity):
    sent = sent.translate(None, string.punctuation)
    ent_item_as_separate_word = False
    # if it appears as separate words inside the text
    if (" " + entity + " ") in sent:
        return True
    # if it appears as separate words at first
    ind_first_match = sent.find(entity+" ")
    if ind_first_match == 0:
        return True
    # if it appears as separate words at the end
    ind_last_match = sent.rfind(" "+entity)
    if ind_last_match + len(entity) + 1 == len(sent):
        return True
    return False
    
```
